```python
#Importing required libraries
import speech_recognition as sr
import pyttsx3
import datetime
import os
import time
from datetime import date


#setup for voice for AI agent
engine=pyttsx3.init('sapi5')
voices=engine.getProperty('voices')
engine.setProperty('voice','voices[0].id')




#Opening text file to journal 
f=open("journal.txt","a")


#Writing date before start of each journal entry
d=date.today()
f.write(str(d)+"\n") 


#function to convert text to speech
def speak(text):
    engine.say(text)
    engine.runAndWait()
    

#function to greet user
def wishMe():
    hour=datetime.datetime.now().hour
    if hour>=0 and hour<12:
        print("So,Hello Anushri,Good Morning")
        speak("So,Hello Anushri,Good Morning")
        print("Start with journal entry")
        speak("Start with journal entry")
    elif hour>=12 and hour<16:
        print("So,Hello Anushri,Good Afternoon")
        speak("So,Hello Anushri,Good Afternoon")
        print("Start with journal entry")
        speak("Start with journal entry")
    else:
        print("So,Hello Anushri,Good Evening")
        speak("So,Hello Anushri,Good Evening")
        print("Start with journal entry")
        speak("Start with journal entry")
        

#Function to take journal input from user
def takeCommand():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        audio=r.record(source,duration=12)
        statement=r.recognize_google(audio)
        print(f"user said:{statement}\n")
        f.write(f"{statement}")
        f.write("\n")
        
#Formal introduction to the user before each entry
print("This is your AI personal assistant journal Piku")
speak("This is your AI personal assistant journal Peeku")
time.sleep(2)
#finding current time
strTime=datetime.datetime.now().strftime("%H:%M")
#to get today's date
d=date.today()
print(f"Today's date is: {d}" )
speak(f"Today's date is: {d}" )
print(f"And the time is {strTime}")
speak(f"And the time is {strTime}")


#greeting call function
wishMe() 


#journal input call function
takeCommand() 


#loop to continue/stop journaling 
while (True):
    speak("Are you still there? Do you want to continue?")
    r=sr.Recognizer()
    with sr.Microphone() as source:
        dn=4
        audio=r.record(source,duration=dn)
        st=r.recognize_google(audio)
        print(f"user said:{st}\n")
        if 'yes' in st:
            speak("You can Continue")
            takeCommand()
        elif 'no' or 'stop' or 'don\'t' or 'bye' or 'good bye' or 'ok bye' in st:
            speak('okay then your personal assistant journal Peeku is shutting down,Good bye')
            f.write("-------------------------------------------------------------------------------\n")
            f.close()
            break
            
            
#Finishing journal entry 
f.write("-------------------------------------------------------------------------------\n")


#closing text file(journal)
f.close()
       
```
